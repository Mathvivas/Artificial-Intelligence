## Questions and Answers Pipeline

```python
qa = pipeline('question-answering', model='pierreguillou/bert-base-cased-squad-v1.1-portuguese')
```

```python
text = 'Na tecnologia, a Inteligência Artificial (de sigla: IA; do inglês: \
Artificial Intelligence, de sigla: AI) é a inteligência demonstrada por \
máquinas ao executar tarefas complexas associadas a seres inteligentes, \
além de também ser um campo de estudo acadêmico, no qual o principal objetivo \
é de executar funções de modo autônomo.'
question = 'Qual é a sigla de Inteligência Artificial?'
answer = qa(question=question, context=text)
print('Question: ', question)
print('Answer: ', answer['answer'])
print('Score: ', answer['score'])
```

```python
question = 'O que é Inteligência Artificial?'
answer = qa(question=question, context=text)
print('Question: ', question)
print('Answer: ', answer['answer'])
print('Score: ', answer['score'])
```

```python
from bertopic import BERTopic
from sklearn.feature_extraction.text import CountVectorizer
from sentence_transformers import SentenceTransformer
```

```python
import glob
```

```python
data = []
dir = '/content/drive/MyDrive/Colab Notebooks/Dados/BERTopicDocs/'
for archive in glob.iglob(dir + '*.txt', recursive=False):
    my_archive = open(archive, 'r', encoding='utf8')
    content = my_archive.read()
    data.append(content)
    my_archive.close()
```

```python
len(data)
```

```python
data[3]
```

```python
sentences = []
for item in data:
    for sentence in sent_tokenize(item):
        sentences.append(sentence)
```

```python
len(sentences)
```

```python
sentences[200]
```

```python
model = BERTopic(language='portuguese', verbose=True, 
                 top_n_words=15, min_topic_size=10, nr_topics=20,
                 vectorizer_model=CountVectorizer(ngram_range=(1, 3),
                                stop_words=stopwords.words('portuguese'),
                                min_df=10))
```

```python
%%time
topics, probabilities = model.fit_transform(sentences)
```

```python
freq = model.get_topic_info()
freq.head(10)
```
